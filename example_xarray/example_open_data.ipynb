# Example loading NetCDF data saved with ArgoBV

This is a *very* quick example of how to load data NETCDF data downloaded from the ArgoBV app.

All of the data retrieved by the query is downloaded to netcdf, even if temporal selection removes some of the queried data.

The argopy point to profile function was applied, so the data is in terms of profiles, not individual measurements.

In order to encode the data for download, the __Fetched_constraints__ attribute is set to a string "none". This is because of an encoding error. Also, the __Fetched_uri__ is set as a string, not as an array.

```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds=xr.open_dataset('example_argo_data.nc')
```

```python
ds
```

```python
ds.plot.scatter(x="LONGITUDE",y="LATITUDE")
```

```python
plt.plot(ds.TEMP[0,:],ds.PRES[0,:])
plt.xlabel("Temperature")
plt.ylabel("Pressure")
plt.ylim([500,0])
plt.grid(alpha=0.2)
plt.show()
```
